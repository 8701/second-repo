```python
q=0.5
R=2.5
def Eptr(n,q,R):
  return (n/(2-2*n))+(q*R*1/2)
from scipy import arange
range_n = arange(0,1.00,0.01)
print (range_n)
range_Eptr = [Eptr(n,q,R) for n in range_n]
print (range_Eptr)
import matplotlib.pyplot as plt
plt.clf()
plt.plot(range_n, range_Eptr,'-', color = 'b', linewidth = 2)
plt.title("$Transparency Expected lifetime value of originating one unit of asset$",fontsize = 14)
plt.xlabel("$n$",fontsize = 14)
plt.ylabel("$Eptransparency$",fontsize = 14)
plt.xlim(0.0,1.0)
plt.ylim(0.0,20.0)
plt.savefig('Transparency_Expected_lifetime_value_of_originating_one_unit_of_asset.png')
```
